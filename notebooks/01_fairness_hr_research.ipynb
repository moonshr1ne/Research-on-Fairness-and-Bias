## Research question
Do HR-related machine learning models show measurable demographic disparities, and can mitigation methods reduce these disparities without unacceptable loss of predictive performance?

## Hypotheses

**H1. Protected-feature removal test.** Removing `Gender`, `Age`, and `AgeGroup` should reduce the Gender DP gap by at least 80%. Observed: the gap grew from 0.069 to 0.141, so H1 is falsified, consistent with strong proxy effects.  
**H2. Age disparity dominance.** AgeGroup DP gap should exceed Gender DP gap by at least 5 percentage points for the baseline model.  
**H3. Mitigation trade-off.** Reweighting should reduce Gender DP gap by at least 25%, with a ROC-AUC or F1 cost.  
**H4. Model complexity trade-off.** A tree-based model should improve ROC-AUC by at least 0.02 compared with Logistic Regression, but not necessarily improve fairness.
